Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

Lendo o csv

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_468477.csv')
```

Mostrando as primeiras linhas e colunas do csv

```python
print(df.head())
```

<h2>Gráfico de Linha<h2>
<h4>Este código gera um gráfico de linha que mostra a evolução da população ao longo de anos selecionados para o Brasil.<h4>

```python
brasil = df[df['country_name'] == 'Brazil']
anos_selecionados = ['1961', '1970', '1980', '1990', '2000', '2010', '2020', '2022']
brasil = brasil.set_index('country_name').T.loc[anos_selecionados]
plt.figure(figsize=(10,6))
plt.plot(brasil.index, brasil['Brazil'], marker='o')
plt.title('População ao longo dos anos - Brasil')
plt.xlabel('Ano')
plt.ylabel('População')
plt.grid(True)
plt.show()
```

<h2>Gráfico de Barra<h2>
<h4>Esse código gera um gráfico de barras que mostra a população para os dez principais países em 2022, selecionando e classificando os dados do DataFrame original para esse ano específico.<h4>

```python
ano_2022 = df[['country_name', '2022']]
ano_2022 = ano_2022.sort_values(by='2022', ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(ano_2022['country_name'], ano_2022['2022'])
plt.title('População em 2022 - Top 10 países')
plt.xlabel('País')
plt.ylabel('Valor da população')
plt.xticks(rotation=45)
plt.show()
```

<h2>Gráfico de Histograma<h2>
<h4>
Este código cria um histograma que mostra a distribuição da população para o ano de 2022, com 20 intervalos representando a frequência de diferentes faixas de população.<h4>

```python
valores_2022 = df['2022']
plt.figure(figsize=(10,6))
plt.hist(valores_2022.dropna(), bins=20, edgecolor='k')
plt.title('Distribuição da População em 2022')
plt.xlabel('População')
plt.ylabel('Frequência')
plt.show()
```

<h2>Gráfico de Pizza<h2>
<h4>Este código seleciona e classifica os dados da população para o ano de 2022 e, em seguida, gera um gráfico de pizza mostrando a proporção da população nos 10 principais países neste ano.<h4>


```python
ano_2022 = df[['country_name', '2022']].dropna()
pais_mais_populoso = ano_2022.loc[ano_2022['2022'].idxmax()]
ano_2022 = ano_2022[ano_2022['country_name'] != pais_mais_populoso['country_name']].sort_values(by='2022', ascending=False).head(9)
top_10_valores = ano_2022['2022'].astype(float)
top_10_paises = ano_2022['country_name']
plt.figure(figsize=(10,6))
plt.pie(top_10_valores, labels=top_10_paises, autopct='%1.1f%%', startangle=140)
plt.title('Proporção dos valores da população em 2022 - Top 10 países (excluindo o mais populoso)')
plt.show()
```

<h2>Gráfico de Scatter<h2>
<h4>
Esse código gera um gráfico de dispersão para comparar os valores de um indicador específico em 2000 e 2020 para os países disponíveis nos dados, mostrando a relação entre esses valores ao longo desses dois anos.<h4>


```python
scatter_data = df[['country_name', '2000', '2020']].dropna()

plt.figure(figsize=(10,6))
plt.scatter(scatter_data['2000'], scatter_data['2020'])
plt.title('Relação entre os valores do Indicador em 2000 e 2020')
plt.xlabel('Valor do Indicador em 2000')
plt.ylabel('Valor do Indicador em 2020')
plt.grid(True)
plt.show()
```
